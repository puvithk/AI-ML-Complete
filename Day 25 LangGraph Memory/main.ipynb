Aritmitic Agent 

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.prebuilt import ToolNode , tools_condition
from langchain.messages import HumanMessage , SystemMessage
from langgraph.graph import START , StateGraph , END , MessagesState 
from IPython.display import display , Image
from pprint import pprint
from dotenv import load_dotenv
from langgraph.checkpoint.memory import MemorySaver

load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(
    model = 'gemini-3-flash-preview'
)
```

Tool for agent 


```python
def multiply(a:int , b:int) -> int:
    """Multiply a and b 
    
    Keyword arguments:
    a -- number 1
    b -- number 2 
    Return: result after multipling 
    """
    return a*b

def add(a:int , b:int) -> int :
    """Add a and b 
    
    Keyword arguments:
    a -- number 1
    b -- number 2
    Return: result after adding 
    """
    return a+b 

    

```

Bind the tools with the llm 

```python
llm_with_tools = llm.bind_tools([multiply , add])
```

State for the graph 

```python
class State(MessagesState):
    pass
```

Tool calling Llm  function


```python
system_message = SystemMessage(content="Your are a assiatnce which can perform arithmetic operations")

def tools_calling_llm(state : State):
    return {
        'messages' : [llm_with_tools.invoke([ system_message ]+ state['messages'])]
    }
```

Build the agent 


```python
builder = StateGraph(State)
builder.add_node("tools_calling_llm" , tools_calling_llm)

builder.add_edge(START , "tools_calling_llm")
builder.add_node('tools' , ToolNode([multiply , add]))
builder.add_conditional_edges(
    "tools_calling_llm" , 
    tools_condition
)
builder.add_edge('tools' , "tools_calling_llm")
builder.add_edge("tools_calling_llm" , END)
```

```python
agent = builder.compile()
```

```python
result = agent.invoke(
    {
        "messages" : HumanMessage("What is 10 add by 2" , name="puvith")
    }
)
```

Adding Memeroy saver based on the thread id


```python
memory = MemorySaver()
react_memory_agent = builder.compile(checkpointer=memory)
```

Update the thread id 

```python
config = {
    "configurable" : {
        "thread_id" : "1"
    }
}
result = react_memory_agent.invoke(
    {
        "messages" : HumanMessage("What is 10 add by 2" , name="puvith")
    } , config
)
```

```python
result = react_memory_agent.invoke(
    {
        "messages" : HumanMessage("For that multipy 6" , name="puvith")
    }, config
)
```
